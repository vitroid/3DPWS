```python
from ClustIce.water import tip4p
from genice2.rigid import euler2rotmat
import numpy as np
import py3Dmol

def load(f):
    cols = f.readline().split()
    ntotal, nw, ng = [int(x) for x in cols[:3]]
    cell = float(cols[3]) / 10

    s = ""
    s += "2gro.py\n"
    s += f"{nw * 4 + ng}\n"

    for i in range(nw):
        six = [float(x) for x in f.readline().split()]
        pos = np.array(six[:3]) / 10
        euler = six[3:]
        rotmat = euler2rotmat(euler)
        sites, names = tip4p()
        water = sites @ rotmat + pos
        for site, name in zip(water, names):
            s +=                 f"{i+1:5d}SOL  {name:5s}99999{site[0]:>8.3f}{site[1]:>8.3f}{site[2]:>8.3f}\n"


    for i in range(ng):
        site = [float(x) / 10 for x in f.readline().split()[:3]]
        name = "Xe"
        s +=             f"{i+1:5d}GAS  {name:5s}99999{site[0]:>8.3f}{site[1]:>8.3f}{site[2]:>8.3f}\n"
    s += f"{cell} {cell} {cell}\n"
    return s

with open("100/fort.100") as f:
    gro = load(f)
view = py3Dmol.view()
view.addModel(gro, "gro")
view.setStyle({"stick": {}})
view.setStyle({'elem': 'Xe'}, {'sphere': {'scale': 0.5, 'color': '#0ff'}})
view.zoomTo()
view.show()
```
